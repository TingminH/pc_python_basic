百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
total = 3000

if total >= 2000:
    total -= 200

print(total)
```

```python
# 進階: 滿 4000 折 400 滿 6000 折 600
# hint:相除取整數 有關

def main():
    # 輸入消費金額
    total = int(input("消費金額: "))
    

    if total >= 6000:  
        total -= 600
        print(f"折抵600")

    if total >= 4000:
        total -= 400
        print(f"折抵400")
    else :
        print(f"無折抵")
        
    print(total)
```
