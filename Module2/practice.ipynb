百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
def main():
    total = int(input("消費金額: "))

    if total >= 2000:
        total -= 200

    print(total)
```

```python
# 老師的code
total = 3000
if total >= 2000:
    total -= 200

print(total)
```

```python
# 進階: 滿 4000 折 400 滿 6000 折 600
# hint:相除取整數 有關

def main():
    # 輸入消費金額 / Enter spending amount
    total = int(input("消費金額: "))
    print("消費金額:", total)
    # 判斷折扣 / Check discount
    if total >= 6000:
        total -= 600
        print("折抵600")
    elif total >= 4000:
        total -= 400
        print("折抵400")
    elif total >= 2000:
        total -= 200
        print("折抵200")
    else:
        print("無折抵")

    # 顯示實付金額 / Show final amount
    print("實付金額:", total)

main()
```

```python
# 進階: 滿 4000 折 400 滿 6000 折 600
# hint:相除取整數 有關

# 老師的code
total = 5500
discount = 200

if total >= 2000:
    n = total // 2000
    total -= discount*n

print(total)
```
